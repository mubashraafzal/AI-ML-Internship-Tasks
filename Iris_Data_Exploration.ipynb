### Importing Libraries

Imported the necessary libraries.
Pandas is used for handling the dataset, seaborn and matplotlib are used for creating graphs and visualizations.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the Dataset

Loaded the Iris dataset using seaborn.
This dataset contains information about different species of iris flowers.

```python
iris = sns.load_dataset('iris')
```

### Inspecting the Dataset

Checked the shape of the dataset to see how many rows and columns it has.
I also printed the column names and displayed the first 5 rows using head() to understand the structure of the data.

```python
print("Shape of dataset:", iris.shape)
print("\nColumn Names:")
print(iris.columns)

print("\nFirst 5 Rows:")
print(iris.head())
```

### Dataset Information

Used the info() function to check the data types of each column and to see if there are any missing values in the dataset.

```python
print("\nDataset Info:")
iris.info()
```

### Summary Statistics

Used describe() to get statistical information like mean, minimum, maximum, and standard deviation of numerical columns.
This helps in understanding the overall distribution of the data.

```python
print("\nSummary Statistics:")
print(iris.describe())
```

### Scatter Plot Visualization

Created a scatter plot to see the relationship between sepal length and sepal width.
Different colors represent different species of iris flowers.

```python
plt.figure()
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

### Histograms

Created histograms to see the distribution of each numerical feature.
This helps in understanding how the values are spread.

```python
iris.hist()
plt.show()
```

### Box Plot for Outliers

Used a box plot to check for outliers and understand the spread of the data.
Box plots show median, quartiles, and extreme values.

```python
plt.figure()
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Features")
plt.show()
```
